### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
salary= int(input("Enter your salary: "))
HRA=salary*0.1
DA=salary*0.05  
PA=salary*0.03
if salary>=0 and salary<500000:
    tax=0
elif salary>=500000 and salary<1000000:
    tax=salary*0.1
elif salary>=1000000 and salary<1500000:
    tax=salary*0.2
else:
    tax=salary*0.3

deducation=HRA+DA+PA+tax
net_salary=salary-deducation
print(f"salary: {salary}")
print(f"salary after deducation: {net_salary}")

````

### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output: 

```bash
120
```

````python
number=int(input("Enter a number: "))
factorial=1
if number<0:
    print("factorial does not exist for negative numbers")
elif number==0:
    print("factorial of 0 is 1")
else:
    for i in range(1,number+1):
        factorial=factorial*i
    print(f"factorial of {number} is {factorial}")
````

### `Problem 9`: Write a program that keeps on accepting a number from the user until the user enters Zero. Display the sum and average of all the numbers.

````python
number=int(input("Enter a number: "))
sum=0
term=0
#exclude 0 from the sum and average calculation
while number!=0:
    sum=sum+number
    term+=1
    avg=sum/term
    number=int(input("Enter a number: "))
print(f"the sum of all input number is: {sum}")
print(f"the average of all input number is: {avg}")
````

````python
### `Problem 12`: A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps. 
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps. 

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.** 

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```
````

````python
x=0
y=0
while True:
    move=input(""" 
    enter the movement of the robot in the format 
               press 1 for up
               press 2 for down
               press 3 for left 
               press 4 for right
               press ! to stop""")
    if move=="1":
        steps=int(input("enter the number of steps: "))
        y+=steps
    elif move=="2":
        steps=int(input("enter the number of steps: "))
        y-=steps    
    elif move=="3":
        steps=int(input("enter the number of steps: "))
        x-=steps
    elif move=="4":
        steps=int(input("enter the number of steps: "))
        x+=steps
    elif move=="!":
        break
distance=(x**2+y**2)**0.5
print(f"the distance from the current position to the original point is: {round(distance)}")
````

###`Problem 14`:Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
#here we consider all one digit number as armstrong number
num=int(input("range to find armstrong number"))
for i in range(0,num+1):
    l=len(str(i))
    c=i
    sum=0
    while i!=0:
        a=i%10
        b=i//10
        sum=sum+a**l
        i=b
    if c==sum:
        print(c,end=" ")
````
