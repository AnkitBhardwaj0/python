### number guessing

```python
import random
lucky_number=random.randint(1,100)
attempt=1
predict=int(input("guess the number"))
while(lucky_number!=predict and attempt<=5):
    if(lucky_number>predict):
        print("guess number is less than luky number")
    else:
        print("guess number is greater than lucky number")
    predict=int(input("guess the number"))
    attempt+=1
if(lucky_number==predict ):
    print("congratulations you won !")
    print(f"number of attempts {attempt}  ")
else:
    print("better luck next time !")
```
