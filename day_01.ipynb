### Q3:- Take 2 numbers as input from the user.Write a program to swap the numbers without using any special python syntax.

````python
num1=int(input('enter first number'))
num2=int(input('enter second number'))
print('Before swapping: num1=',num1,' num2=',num2)
#swapping logic without using third variable
num1=num1+num2
num2=num1-num2
num1=num1-num2
print('After swapping: num1=',num1,' num2=',num2)
num1,num2= num2,num1
print('After swapping using tuple: num1=',num1,' num2=',num2)
num1=num1+num2 - (num2:=num1)
print('After swapping using walrus operator: num1=',num1,' num2=',num2)
````

### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
print('enter first coordinate (x1,y1)')
x1=int(input('enter x1'))
y1=int(input('enter y1'))
print('You entered first coordinate as (',x1,',',y1,')')
print('enter second coordinate (x2,y2)')
x2=int(input('enter x2'))
y2=int(input('enter y2'))
print('You entered second coordinate as (',x2,',',y2,')')
distance=((x2-x1)**2 + (y2-y1)**2)**0.5
print('Distance between points (',x1,',',y1,') and (',x2,',',y2,') is:',distance)
````

### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2

````python
legs=int(input('enter number of legs'))
heads=int(input('enter number of heads'))
dogs=(legs-2*heads)//2
chickens=heads-dogs
print('Number of dogs:',dogs)
print('Number of chickens:',chickens)
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
first_term=int(input('enter first term of AP'))
second_term=int(input('enter second term of AP'))
diff=second_term-first_term
n=int(input('enter the position of the term you want to find'))
n_th_term=(n-1)*diff + first_term
print('n-th term of AP is:',n_th_term)
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.

Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
print('enter the dimension of milk tank')
height_t=float(input('enter height of milk tank'))
width_t=float(input('enter width of milk tank'))
length_t=float(input('enter length of milk tank'))
volume_of_milk_tank=height_t*width_t*length_t
print('Volume of milk tank is:',volume_of_milk_tank)
print('enter the dimension of glass')
height=float(input('enter height of glass'))
radius=float(input('enter radius of glass'))
volume_of_glass=3.14*radius**2*height
print('Volume of glass is:',volume_of_glass)
number_of_glasses=volume_of_milk_tank//volume_of_glass
print('Number of glasses required to fill the milk tank is:',number_of_glasses)

````
